```python
import pandas as pd

data_1 = pd.read_csv('data/dataset_export_effectifs.csv')
data_2 = pd.read_csv('data/dataset_export_demographie-effectifs-et-les-densites.csv')
data_3 = pd.read_csv('data/dataset_export_prescriptions.csv')
data_4 = pd.read_csv('data/dataset_export_honoraires-detailles.csv')
```

```python
def standardize_columns(df):
    df['profession_sante'] = df['profession_sante'].astype(str).str.strip()
    df['departement'] = df['departement'].astype(str).str.strip()
    df['region'] = df['region'].astype(str).str.strip()
    df['libelle_region'] = df['libelle_region'].astype(str).str.strip()
    df['libelle_departement'] = df['libelle_departement'].astype(str).str.strip()
    return df

data_4 = data_4.loc[data_4['annee'] > 2022, :]
data_3 = data_3.loc[data_3['annee'] > 2022, :]
data_2 = data_2.loc[data_2['annee'] > 2022, :]
print(len(data_4))

data_4 = standardize_columns(data_4)
data_3 = standardize_columns(data_3)
data_2 = standardize_columns(data_2)
print("column standardized")
# 2. Merge Datasets
# Merge data_4 (honoraires) and data_3 (prescriptions)
merged_df = data_4.merge(
    data_3,
    on=['annee', 'profession_sante', 'departement', 'region', 'libelle_region', 'libelle_departement'],
    how='inner',  # Use 'left' if you want to keep all rows from data_4
    suffixes=('_honoraires', '_prescriptions')
)

# Merge the result with data_2 (effectif and densit)
merged_df = merged_df.merge(
    data_2,
    on=['annee', 'profession_sante', 'departement', 'region', 'libelle_region', 'libelle_departement'],
    how='inner',  # Use 'left' if you want to keep all rows from the previous merge
    suffixes=('', '_effectif')
)

# 3. Handle Missing Data
# Fill missing values with 0 where appropriate (e.g., numeric columns)
numeric_columns = merged_df.select_dtypes(include=['float64', 'int64']).columns
merged_df[numeric_columns] = merged_df[numeric_columns].fillna(0)

# Fill missing categorical data with 'Unknown' if necessary
categorical_columns = merged_df.select_dtypes(include=['object']).columns
merged_df[categorical_columns] = merged_df[categorical_columns].fillna('Unknown')
```
